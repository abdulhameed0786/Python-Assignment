# Qno.1: Define Object Oriented Programming Language?

# Ans:
    Object Oriented Programming (OOP) refers to a type of computer programming (software design) in which programmers define the data type of a data structure, and also the types of operations (functions) that can be applied to the data structure.
     In this way, the data structure becomes an object that includes both data and functions. In addition, programmers can create relationships between one object and another. For example, objects can inherit characteristics from other objects.

### It has four main pillars that are given bellow:

## 1.Inheritance:
         A feature that represents the "is a" relationship between different classes.

## 2.Encapsulation:
         The process of combining elements to create a new entity. A procedure is a type of encapsulation because it combines a series of computer instructions.

## 3.Abstraction:
         The process of picking out (abstracting) common features of objects and procedures.

## 4.Polymorphism:
         A programming language's ability to process objects differently depending on their data type or class.

# Qno.3: Differentiate between function and method?
# Ans: 
### A function is a piece of code that is called by name. It can be passed data to operate on (i.e. the parameters) and can optionally return data (the return value). All data that is passed to a function is explicitly passed.
### A method is a piece of code that is called by a name that is associated with an object. In most respects it is identical to a function except for two key differences:
    1.	A method is implicitly passed the object on which it was called.
    2.	A method is able to operate on data that is contained within the class (remembering that an object is an instance of a class - the class is the definition, the object is an instance of that data).

## 2.	Object:

        An object (instance) is an instantiation of a class. When class is defined, only the description for the object is defined. Therefore, no memory or storage is allocated.

#### The example for object of parrot class can be:

 		Obj = Parrot()
        
    Here, obj is object of class Parrot.

## 3.	Attribute:
### There are two types of Attributes:
### a) Instance attribute:
        An instance attribute is a Python variable belonging to one, and only one, object. This variable is only accessible in the scope of this object and it is defined inside the constructor function, __init__(self,…), of the class.
### b) Class attribute:
        A class attribute is a Python variable that belongs to a class rather than a particular object. It is shared between all the objects of this class and it is defined outside the constructor function, __init__(self,…), of the class.
#### For Example:-
            class Example(object):
                class_attr = 0
                def__init__ (self, instance_attr):
                    self. instance_attr = instance_attr
•	instance_attr is an instance attribute defined inside the constructor.
•	class_attr is a class attribute defined outside the constructor.

```python
# Question no: 5 
class Car:
    #instantiation method
    def __init__(self,name,model,color,driver,price):
        self.name = name
        self.model = model
        self.color = color
        self.driver = driver
        self.price = price
    # First method to update driver name
    def Up_Driver(self,NewDriver):
        self.driver = NewDriver
    
    # Second method to update model
    def Up_Model(self,NewModel):
        self.model = NewModel
    
    # Third method to update color of car 
    def Up_Color(self,NewColor):
        self.color = NewColor
# First Object instance        
car1 = Car("Lexus LX", 2018,"Silver","John Doe","$99,364")
# Second Object instance
car2 = Car("BMW 750i xDrive", 2018,"Grey","Baraq Ubama","$99,400")
# Third Object instance
car3 = Car("Land Rover Range Rover Velar", 2018, "Red","John Doe","$98,495")
# Fourth Object instance
car4 = Car("Cadillac Escalade ESV Platinum 4WD", 2018,"Silver","John Doe","$99,795")
# Fifth Object instance
car5 = Car("Jaguar F-Type R", 2018,"Silvery White","John Doe","$99,900")


print("->First Car")
print("Car Name: ",car1.name,"Car Model: ", car1.model,"\nCar Color: ", car1.color,"\nCar Driver Name: ", car1.driver,"\nCar Price: ", car1.price)
print("\n->Second Car")
print("Car Name: ",car2.name,"Car Model: ", car2.model,"\nCar Color: ", car2.color,"\nCar Driver Name: ", car2.driver,"\nCar Price: ", car2.price)
print("\n->Third Car")
print("Car Name: ",car3.name,"Car Model: ", car3.model,"\nCar Color: ", car3.color,"\nCar Driver Name: ", car3.driver,"\nCar Price: ", car3.price)
print("\n->Fourth Car")
print("Car Name: ",car4.name,"Car Model: ", car4.model,"\nCar Color: ", car4.color,"\nCar Driver Name: ", car4.driver,"\nCar Price: ", car4.price)
print("\n->Fifth Car")
print("Car Name: ",car5.name,"Car Model: ", car5.model,"\nCar Color: ", car5.color,"\nCar Driver Name: ", car5.driver,"\nCar Price: ", car5.price)


print("\n")
print("First Car's Old Driver, Model and Color \n")
# old attributes
print("Old Driver:",car1.driver)
print("Old Model:",car1.model)
print("Old Color:",car1.color)

print("\n")
# updating attributes
car1.Up_Driver("Johnson")
car1.Up_Model(1999)
car1.Up_Color("Green")

print("First Car's Updated Driver, Model and Color \n")
# updated attributes
print("Updated Driver:",car1.driver)
print("Updated Model:",car1.model)
print("Updated Color:",car1.color)

print("\n")
```
